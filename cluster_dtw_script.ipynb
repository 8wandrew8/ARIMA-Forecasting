```python
from sklearn import preprocessing
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
plt.rcParams['figure.figsize'] = 15, 6 #CHECK
```

```python
tickers_list = pd.read_html('https://vseakcii.ru/issuers/countries/Russia?limit=-1')[0]
tickers_list = tickers_list[['Тикер', 'Название', 'Сектор']]
tickers_list.index = tickers_list['Тикер']
tickers_list.pop('Тикер')
```

```python
tickers_list
```

```python
#DOWNLOAD DATA OF TICKERS MOEX
import requests
import apimoex
import sys
t = pd.DataFrame(apimoex.get_board_candles(requests.Session(), security="QIWI", interval=24, start="2021-09-01", end='2022-02-20', board="TQBR"))

process = 0 #just for visualization of downloading process
#data = pd.DataFrame(columns = t['begin']) #columns as levels of time series
#data["for_ts"] = [i for i in range(1, 90)] #just to insert ts row-wisely
#data.index = tickers_list['Тикер']
data_viz = pd.DataFrame() #for visualization

with requests.Session() as session:
    for i in range(len(tickers_list)):
        process = process + 1
        print((process / len(tickers_list)) * 100, '%', process)
        temp = apimoex.get_board_candles(session, security=tickers_list["Тикер"][i], interval=24, start="2021-09-01", end='2022-02-20', board="TQBR")
        if temp == []:
            continue
        temp = pd.DataFrame(temp)
        df = pd.DataFrame(temp['close']) #obtain close prices
        try:
            #data.loc[(data['for_ts'] == process)] = tuple(df['close']) + (process,)
            data_viz[tickers_list["Тикер"][i]] = df
        finally:
            continue
            
#data.pop('for_ts')
```

```python
data_scaled=(data_viz-data_viz.mean())/data_viz.std()
data_scaled.plot(legend = False)
```

```python
data_scaled = data_scaled.transpose()
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
data_fin = data_scaled.join(tickers_list, how = 'inner')
data_fin = data_fin.dropna()
data_scaled = data_scaled.dropna()
data_fin = data_fin[['Название', 'Сектор']]
#data_scaled.head(89)
```

```python
#DTW CLUSTERING
import tslearn.clustering as ts

errors = []

for i in range(1, 20):
    print(str(i/19*100) + '%', end = ' ')
    kmeans = ts.TimeSeriesKMeans(n_clusters = i, metric = 'dtw').fit(data_scaled)
    errors.append(kmeans.inertia_)
    
#errors = pd.Series(errors)
```

```python
plt.title('Elbow Chart')
plt.plot([i for i in range(1,20)], errors, label = 'SSE', marker = 'o')
```

```python
kmeans = ts.TimeSeriesKMeans(n_clusters = 4).fit(data_scaled)
```

```python
for n in range(4):
    plt.plot(kmeans.cluster_centers_[n, :, 0].T, label = n)
plt.title("Centroids")
plt.legend()
plt.show()
```

```python
data_fin['cluster'] = kmeans.labels_
data_scaled['cluster'] = kmeans.labels_
```

```python
#INTERPRITATION IS JUST FINE
pd.DataFrame(data_fin.groupby(['cluster'])['Сектор'].value_counts())
```

```python
def plot_cluster_tickers(current_cluster):
    fig, ax = plt.subplots(
        int(np.ceil(current_cluster.shape[0]/4)),
        4,
        figsize=(15, 3*int(np.ceil(current_cluster.shape[0]/4)))
    )
    fig.autofmt_xdate(rotation=45)
    ax = ax.reshape(-1)

    for index, (_, row) in enumerate(current_cluster.iterrows()):
        ax[index].plot(row[0:-3])
        ax[index].set_title(f"{row.Название}\n{row.Сектор}")
        plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
    
t = data_scaled.join(tickers_list)
for cluster in range(4):
    print("=================================================================================")
    print(f"                          Cluster number: {cluster}")
    print("=================================================================================")
    plot_cluster_tickers(t[t.cluster==cluster])
```

```python
#MULTI DIMENSIONAL SCALING
from sklearn.decomposition import PCA

data_pca = data_scaled.drop('cluster', axis = 1)
pca = PCA(n_components=4)

pca_transformed = pca.fit_transform(data_pca)

plt.figure(figsize=(10, 10))
plt.scatter(pca_transformed[:, 0], pca_transformed[:, 1], c = data_scaled.cluster)
plt.title("PCA: first two pcs")
plt.show()
```
